# Import des données

```python
import pandas as pd
import numpy as np
df = pd.read_csv("salary1.csv")
df.columns=["age","statut","f3","etudes","nombre_annees_etudes","situation_familiale1","secteur_activite","situation_familiale2","ethnie","genre", "plus-value","moins-value", "nb_h_hebdo","pays_naissance","salary"]

df
```

```python
#On supprime la colonne f3 qui est non informative
del df['f3']
```

On observe que la classe salary est composée de 2 valeurs : salaires inférieurs à 50K et supérieurs à 50K.

```python
#Avant d'effectuer une visualisation des données, on séparer notre base en appretissage/test

from sklearn.model_selection import train_test_split

#On récupère nos X et nos y
X = df.iloc[:,0:13]
y = df.iloc[:,13:14]

#On sépare nos données en train et en test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

data_train_df = (X_train,y_train)

data_test_df = (X_test, y_test)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing

# transformation de la classe avec le labelEncoder
y_train = y_train.apply(LabelEncoder().fit_transform)
y_train
```

On applique le one-hot encoding sur les variables qualitatives nominales de manière à les transformer en variables quantitatives.

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['statut'], prefix='statut')], axis=1)
```

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['situation_familiale1'], prefix='sf1')], axis=1)
```

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['secteur_activite'], prefix='se')], axis=1)
```

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['situation_familiale2'], prefix='sf2')], axis=1)
```

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['ethnie'], prefix='et')], axis=1)
```

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['genre'], prefix='s')], axis=1)
```

```python
X_train = pd.concat([X_train, pd.get_dummies(X_train['pays_naissance'], prefix='pn')], axis=1)
```
